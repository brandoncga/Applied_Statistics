## Los resultados ENARM 2019 ya se publicaron, supongamos que la ultraderecha acusa al PG de bajar la dificultad del la cubrir la demanda de doctores. Mediante el RANGO DE SIGNIFICANCIA valide o rechace la acusación

```r
Historico <- read.csv("enarm.csv")
```

```r
Enarm2019 <- read.csv("20199.csv")
```

```r
efihisto <- Historico$SELECCIONADOS / Historico$SUSTENTANTES
```

```r
efi2019 <- Enarm2019$SELECCIONADOS / Enarm2019$SUSTENTANTES
```

```r
options(repr.plot.width=12,repr.plot.height=12/1.61803)
hist (efihisto, main="Frecuencia de Eficiencias históricas del ENARM ",
      col="dark blue", border="light blue", labels=T, xlab="Eficiencia", ylab="FRECUENCIAS",ylim=c(0,500))
```

```r
options(repr.plot.width=12,repr.plot.height=12/1.61803)
hist (efi2019, main="Frecuencia de Eficiencias del ENARM 2019",
      col="dark blue", border="light blue", labels=T, xlab="Eficiencia", ylab="FRECUENCIAS",ylim=c(0,50))
```

# Calculando el rango de significancia

$$rango = \mu \pm 2 \sigma $$

```r
media <- mean (efihisto,na.rm = T)
```

```r
Desviacion <- sd(efihisto,na.rm=T)
```

```r
2*Desviacion
```

```r
media + (2*Desviacion)
media - (2*Desviacion)
```

Calculando la media de las eficiencias de 2019

```r
mean (efi2019,na.rm=T)
```

## La media está dentro del rango de significancia, podemos decir que el PG no influyó en el acceptance rate de este año
